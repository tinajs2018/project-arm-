```python
import pandas as pd
import statsmodels.api as sm
data = pd.read_csv('data - Sheet1.csv')
# Assuming your data is in a CSV file named "your_data.csv"
try:
  # Load data from CSV with header row (adjust as needed)
  data = pd.read_csv("data - Sheet1.csv", header=0)
except FileNotFoundError:
  print("Error: Data file 'your_data.csv' not found. Please check the file path.")
  exit()

# Define variables based on column names (replace with actual names)
dependent_variable = "Country"
independent_variables = ["Inflation", "Exchange Rate", "Government Debt", "Government Revenue"]

# Check for missing values (optional)
if data.isnull().values.any():
  print("Warning: Your data contains missing values. Consider handling them (e.g., imputation) before proceeding.")

try:
  # Select relevant data (handle potential IndexError)
  gdp_usd = data[dependent_variable].astype(float).values
  inflation = data[independent_variables[0]].astype(float).values
  exchange_rate = data[independent_variables[1]].astype(float).values
  govt_debt = data[independent_variables[2]].astype(float).values
  govt_revenue = data[independent_variables[3]].astype(float).values

except IndexError:
  print("Error: Indexing issue with the DataFrame. Check column names and data structure.")
  exit()

# Create the model and fit the regression
model = sm.OLS(gdp_usd, sm.add_constant(np.column_stack([inflation, exchange_rate, govt_debt, govt_revenue])))
results = model.fit()

# Print the summary of the regression analysis
print(results.summary())
```
